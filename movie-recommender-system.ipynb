```python
# import libraries
import numpy as np
import pandas as pd
```

```python
# creating dataframe
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
# viewing single movie info
movies.head(1)
```

```python
# viewing single credit info
credits.head(1)   
```

```python
# merge two dataframes   -- since merged on the basics of title so 1 col less
movies = movies.merge(credits,on="title")
movies.head(1)
```

```python
# removing irrelevant columns
# used columns - genres, id, keywords, title, overview, cast, crew
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
# preprocessing data format

# removing missing data
movies.isnull().sum()
```

```python
# drop null values
movies.dropna(inplace=True)
movies.isnull().sum()
```

```python
# check if duplicate data present
movies.duplicated().sum()
```

```python
# format genre column in right format
genres = movies.iloc[0].genres
```

```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# format as - ['Action','Adventure', 'Fantasy']
```

```python
import ast
ast.literal_eval(genres)

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
# taking top 3 cast names
def convert1(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert1)
movies.head()
```

```python
# taking director name from cast column
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()
```

```python
# converting overview column's string to list format
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
# merge all lists to one list and convert to string - paragraph

# apply genres, keywords, cast, crew a transformation that remove spaces from names for tag formation
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies.head()
```

```python
# merge columns to form tags column
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
# deleting rest of the columns
# new dataframe creation
new_df = movies[['movie_id','title','tags']]
new_df
```

```python
# tags column's data convert list to string
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
new_df.head()
```

```python
new_df['tags'][0]
```

```python
# turn everthing in lowercase
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
new_df.head()
```

```python
# text vectorization - convert the tags into vector to find similarities btw movies
# technique used - bag of words (combine all tags, analyse most used tag )
# sklearn CountVectorizer used 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
vectors[0]
```

```python
len(cv.get_feature_names_out())
```

```python
# used stemming #'loved','loving','loves' => 'love' 'love' 'love'
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
stem('In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. Action Adventure Fantasy ScienceFiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d SamWorthington ZoeSaldana SigourneyWeaver JamesCameron')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
new_df.head()
```

```python
# calculating cosine distance 
# the far distance weaker recommendation

from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]  # fetching the similarity of 2nd movie with other movies
```

```python
# fetch index of movie
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
